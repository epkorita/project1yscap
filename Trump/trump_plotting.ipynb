```python
%matplotlib inline
import os 
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import numpy as np
import datetime as dt
import pprint as pprint


trump_data = './Trump_final.csv'
data = pd.read_csv(trump_data)
df = data.drop(['Unnamed: 0'], axis = 1)
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%m/%d')
df = df.set_index("Date")
df
```

```python
price = df["SPX Open Price"]
volume = df["SPX Volume"]
vs_all = df.iloc[:,2:6]

# plotting the graph by using object oriented plot
fig =plt.figure(figsize=(16,12))    # creating figure object
ax = fig.add_subplot(212)           # adding axes on this figure

# instantiate a second axes that shares the same x-axis
ax3 = ax.twinx()

# ploting Stock price and Vader Score on the same axis
price.plot(ax=ax, label="SPX Price",color="b", marker="o", markersize=10, lw=5,ls='--', alpha=1)
vs_all.plot(ax=ax3, label="Vader Score",color=["r","g","y"],markersize=10, lw=4,ls='-',alpha=1)


# designing labile and ticks
ax.set_ylabel("Stock Price",fontdict = {"fontsize" : 22, "fontweight": "bold"})
ax.set_xlabel("Date",fontdict = {"fontsize" : 22, "fontweight": "bold"})
ax3.set_ylabel("Trump Vader Score",fontdict = {"fontsize" : 22, "fontweight": "bold"})
ax.set_xticklabels(df.index, fontdict = {"fontsize" : 12, "fontweight": "bold"}, rotation = 75)
ax.set_xticks(np.arange(0, 20))

# setting a title
ax.set_title("SPX Price vs Trump Vader Score", fontdict = {"fontsize" : 28, "fontweight": "bold"})
ax.legend(numpoints = 2, frameon = True, markerscale = 1.5, edgecolor = 'blue', fontsize = '12', framealpha = 1, loc="upper center")
ax.grid()
# fig.savefig('trump_price.png', bbox_inches='tight')
plt.show()
```
